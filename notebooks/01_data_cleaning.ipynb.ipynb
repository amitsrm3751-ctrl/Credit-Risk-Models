```python
# 1. Flag missing income
df['MonthlyIncome_missing'] = df['MonthlyIncome'].isnull().astype(int)

# 2. Fill missing values
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())

```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
## 1. Missing Flag
df['Dependents_missing'] = df['NumberOfDependents'].isnull().astype(int)
# Fill using mode
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(df['NumberOfDependents'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
## Check distribution of RevolvingUtilizationOfUnsecuredLines for any outliers value should be between 0 to 1
df['RevolvingUtilizationOfUnsecuredLines'].describe()
```

```python
df['RevolvingUtilizationOfUnsecuredLines'] = df['RevolvingUtilizationOfUnsecuredLines'].clip(0, 1)
```

```python
df['High_Delinquency_Flag'] = (
    (df['NumberOfTime30-59DaysPastDueNotWorse'] > 10) |
    (df['NumberOfTime60-89DaysPastDueNotWorse'] > 10) |
    (df['NumberOfTimes90DaysLate'] > 10)
).astype(int)
```

```python
df['NumberOfTime30-59DaysPastDueNotWorse'].describe()
```

```python
# Max value 98 and other such data like 50, 45 doesn't seem realistic, it is not possible to be late 98 times in 2 years
df['NumberOfTime30-59DaysPastDueNotWorse'] = df['NumberOfTime30-59DaysPastDueNotWorse'].clip(0,10)
```

```python
## Add the outlier flag
cap = df['DebtRatio'].quantile(0.99)
cap
df['DebtRatio_high_flag'] = (df['DebtRatio'] >= cap).astype(int)
```
